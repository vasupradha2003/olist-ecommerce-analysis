```python
# Remove full duplicate rows 
customers.drop_duplicates(inplace=True)
```

```python
# Standardise string columns 
customers['customer_city']  = (customers['customer_city'].str.strip().str.lower().str.title())
```

```python
customers['customer_state'] = (customers['customer_state'].str.strip().str.upper())
```

```python
# Validate customer_state
# Brazil has 27 states (26 + DF)
valid_states = [
    'AC','AL','AP','AM','BA','CE','DF','ES','GO',
    'MA','MT','MS','MG','PA','PB','PR','PE','PI',
    'RJ','RN','RS','RO','RR','SC','SP','SE','TO']
```

```python
invalid_states = customers[~customers['customer_state'].isin(valid_states)]
```

```python
print(f"\n  Rows with invalid state code : {len(invalid_states)}")
if len(invalid_states) > 0:
    print(invalid_states[['customer_state']].value_counts())
```

```python
# Validate zip code 
# Brazilian zip codes are 5-digit numbers
customers['zip_valid'] = (customers['customer_zip_code_prefix'].astype(str).str.match(r'^\d{5}$'))
```

```python
print(f"\n  Invalid zip codes : " f"{(~customers['zip_valid']).sum()}")
```
